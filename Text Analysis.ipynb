```python
from collections import Counter

def count_words(text):
'''
cette fonction effectue le compte des mots dans le livre.
les caracteres de la liste skip sont remplacer par "",
qui seront par la suite supprime pour qu'il ne soit pas compte comme un mot par le programme.
les mots seront separes par un espace " "
'''
    text = text.lower()
    # Liste de caractere a ne pas prendre en compte lors du comptage de mots
    skips = [".", ",", ":", ";", "'", '"',"[","]","!","?",")","(","-"]  
    for ch in skips:  
        text = text.replace(ch, "")  
    word_counts = Counter(text.split(" "))
    del word_counts['']
    return word_counts
```

Nous allons créer une fonction appelée read_book () qui lira nos livres en Python et les enregistrera comme une longue chaîne dans une variable et la renverra.Le paramètre de la fonction sera l'emplacement du book.txt à lire et sera transmis lors de l'appel de la fonction.


```python
def read_book(title_path): 
    with open(title_path, "r", encoding ="utf8") as current_file: 
        text = current_file.read() 
        text = text.replace("\n", "").replace("\r", "") 
    return text 
```

Total des mots uniques: Nous allons concevoir une autre fonction appelée word_stats (), qui prendra le dictionnaire de fréquence de mot (sortie de count_words_fast () / count_words ()) comme paramètre La fonction retournera le nombre total de mots uniques (somme / total des clés dans le dictionnaire de fréquence des mots) et un dict_values contenant le nombre total d'entre eux ensemble, sous forme de tuple.

```python

def word_stats(word_counts): 
    num_unique = len(word_counts) 
    counts = word_counts.values()
    return (num_unique, counts)
```

```python
'''
lecture d'un livre pour se rassurer que nos fonction fonctionne bien.
'''
#lecture du contenu de Romeo and Juliet.txt
text = read_book("./books/English/shakespeare/Romeo and Juliet.txt") 

#quelques statistique sur ce livre 

word_counts = count_words(text)
(num_unique, counts) = word_stats(word_counts) 
print('Le livre contient : ',num_unique, ' mots.')
print('Le livre contient : ', sum(counts),' mots uniques.')
print('Frequence de chaque mots uniques : ')
print(word_counts)

```

```python
def most_frequent(word_counts): 
    '''
    retourne le top 5 des mots les plus frequents.
    '''
    occurence_count = word_counts 
    return occurence_count.most_common(5)
print('TOP 5 des mots les plus frequents : ')
print(most_frequent(word_counts))
```

```python
def unique_word(text):
    '''
    retourne une liste de mots uniques venant du livre en lecture.
    tout le texte est mis en minuscule afin de comparer tout les mots et en retirer que les valeures uniques.
    '''
    text = text.lower()  
    skips = [".", ",", ":", ";", "'", '"',"[","]","!","?",")","(","-"]
    for ch in skips:  
        text = text.replace(ch," ")
    wordlist = set(text.split(" "))
    wordlist.remove('')
    return list(wordlist)

uniq_wordlist=unique_word(text)
print(uniq_wordlist)
```

```python
def findShortest(alist):
    '''
    retourne le mot le plus cours du livre.
    utilise min() avec comme argument la liste 
    et comme cle de comparaison la longueur des elements de la liste.
    '''
    min_list = min(alist, key=len)
    return (min_list)

def findLongest(alist):
    '''
    retourne le mot le plus cours du livre.
    utilise max() avec comme argument la liste 
    et comme cle de comparaison la longueur des elements de la liste.
    '''
    max_list = max(alist, key=len)
    return max_list
# teste des fonctions , si les mots sont mal separes il y a risque que des erreurs s'y retrouve. 
short_w = findShortest(uniq_wordlist)
long_w = findLongest(uniq_wordlist)

print(long_w , len(long_w))
print(short_w, len(short_w))

```

```python
def count_phr(text) :
    '''
    effectue le compte des phrases , de la meme maniere qu'avec les mots
    '''
    sentences = text.split(".")
    #sentences.remove(" ")
    return len(sentences)

def phrase_list(text) :
    '''
    decompose le fichier text en liste de phrases
    '''
    sentences = text.split(".")
    #sentences.remove(" ")
    return (sentences)

phrases = phrase_list(text)

print('Le livre contient approximativement : ',count_phr(text) ,'phrase(s)' )

print('La phrase la plus courte est : ',findShortest(phrases))

print('La phrase la plus longue est : ',findLongest(phrases))
```
